```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3,3)
qc.h(0)
qc.y(1)
qc.h(2)
qc.cx(0,1)
qc.ccx(0,1,2)
qc.z(1)
from numpy import pi 
qc.rz(pi/5,2)
qc.draw()
```

```python
qc_1 = qc.decompose()                            
qc_1.draw()
```

```python
qc_2 = qc_1.decompose()
qc_2.draw()
```

```python
qc_3 = qc_2.decompose()
qc_3.draw()
```

```python
qc_4 = qc_3.decompose()
qc_4.draw()
```

```python
qc_5 = qc_4.decompose()
qc_5.draw()
```
